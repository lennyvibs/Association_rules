Importing the required libraries

```python
import pandas as pd
from apyori import apriori
```

Loading the dataset using pandas library.

```python
df0=pd.read_csv(r"D:\360Digitmg\ASSIGNMENTS\Ass9\transactions_retail1.csv",header=None)
df=df0.copy()
df.head()
```

Removing the NaN values from all the rows using the fillna function. 

```python
df.fillna('',inplace=True)
df.head()
```

Removing the double quotes from each cell using the strip in lambda function. 

```python
df = df.applymap(lambda x: x.strip("''"))
df.head()
```

Using the shape function to get the no of rows and columns of the dataset

```python
df.shape
```

Converting the dataframe into list by storing it into records.

```python
records = []
for i in range(0, 557041):
    records.append([str(df.values[i,j]) for j in range(0, 6)])
```

```python
records
```

Joining the words in every row together to form a single object. 

```python
record_list=[]
for items in records:
    s=' '.join(items)
    record_list.append(s)
```

```python
record_list
```

Counter is an unordered collection where elements are stored as Dict keys and their count as dict value.

```python
from collections import Counter
item_frequencies=Counter(record_list)
item_frequencies
```

The items in the item_frequencies are sorted on the basis of key value which is the count of items.

```python
item_frequencies=sorted(item_frequencies.items(),key=lambda x:x[1])
item_frequencies
```

Storing frequencies and items in separate variables reversing the order.

```python
frequencies= list(reversed([i[1] for i in item_frequencies]))
items= list(reversed([i[0] for i in item_frequencies]))
frequencies
```

```python
items
```

Using the barplot to show the items which are bought the most using the variables frequencies and items. 

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(15,7))
plt.bar(height=frequencies[0:11],x=list(range(0,11)), color=['crimson','lawngreen','darksalmon','black','yellow','aqua','blueviolet','brown','orange','coral','pink'])
plt.xticks(list(range(0,11),),items[0:11],rotation=90)
plt.xlabel('items')
plt.ylabel('Count')
plt.show()
```

```python
recordlists=[]
for i in record_list:
    recordlists.append([i])
```

```python
recordlists
```

Using the apriori function from the apyori library to get the support confidence and lift values of the itemsets. 

```python
association_ruless = apriori(recordlists, min_support=0.0025, min_lift=1, min_length=1)
association_results = list(association_ruless)
```

```python
for item in association_results:

    # first index of the inner list
    # Contains base item and add item
    pair = item[0] 
    items = [x for x in pair]
    print("Itemsets : "+ items[0])

    #second index of the inner list
    print("Support: " + str(item[1]))

    #third index of the list located at 0th
    #of the third index of the inner list

    print("Confidence: " + str(item[2][0][2]))
    print("Lift: " + str(item[2][0][3]))
    print("=====================================")
```
