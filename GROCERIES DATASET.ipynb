Importing the required libraries

```python
import pandas as pd
from mlxtend.frequent_patterns import apriori, association_rules
from termcolor import colored
```

Loading the file to groceries as a list using open function. 

```python
groceries=[]
with open(r"D:\360Digitmg\ASSIGNMENTS\Ass9\groceries.csv") as f:
    groceries=f.read()
groceries
```

Splitting the data into separate transactions using separator as "\n"

```python
groceries=groceries.split('\n')
groceries
```

create an empty list called groceries list and add the separate lines as separate lists within a list. 

```python
groceries_list=[]
for i in groceries:
    groceries_list.append(i.split(","))
groceries_list
```

The separate lists are declared as items and each product in the items is stored in a variable called all_groceries_list. 

```python
all_groceries_list=[i for item in groceries_list for i in item]
all_groceries_list
```

Counter is an unordered collection where elements are stored as Dict keys and their count as dict value.

```python
from collections import Counter
item_frequencies=Counter(all_groceries_list)
item_frequencies
```

The items in the item_frequencies are sorted on the basis of key value which is the count of items. 

```python
item_frequencies=sorted(item_frequencies.items(),key=lambda x:x[1])
item_frequencies
```

Storing frequencies and items in separate variables reversing the order. 

```python
frequencies= list(reversed([i[1] for i in item_frequencies]))
items= list(reversed([i[0] for i in item_frequencies]))
frequencies
```

```python
items
```

Using the barplot to show the items which are bought the most using the variables frequencies and items. 

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(15,7))
plt.bar(height=frequencies[0:11],x=list(range(0,11)), color=['crimson','lawngreen','darksalmon','black','yellow','aqua','blueviolet','brown','orange','coral','pink'])
plt.xticks(list(range(0,11),),items[0:11],rotation=20)
plt.xlabel('items')
plt.ylabel('Count')
plt.show()
```

Using the TransactionEncoder from mlxtend library to do one hot encoding of the different items.

```python
from mlxtend.preprocessing import TransactionEncoder
te=TransactionEncoder()
x=te.fit_transform(groceries_list)
dfx=pd.DataFrame(x,columns=te.columns_)
dfx.head()
```

The first column in the above table doesn't have name, this represents an empty item which was created while making the list.This is removed using the drop function.

```python
dfx.drop(dfx.columns[[0]],axis = 1,inplace=True)
dfx
```

Using the apriori algorithm to find the most frequent itemsets. 

```python
frequent_itemsets=apriori(dfx, min_support=0.0075,max_len=4, use_colnames=True)
frequent_itemsets.sort_values('support',ascending=False,inplace=True)
frequent_itemsets
```

The most frequent itemsets are plotted as a barplot. 

```python
plt.figure(figsize=(15,7))
plt.bar(x=list(range(0,11)),height=frequent_itemsets.support[0:11],color=['crimson','lawngreen','darksalmon','black','yellow','aqua','blueviolet','brown','orange','coral','pink'])
plt.xticks(list(range(0,11)),frequent_itemsets.itemsets[0:11],rotation=90)
plt.xlabel('item-sets')
plt.ylabel('support')
plt.show()
```

Using the association rules from mlxtend to get the support, confidence and lift values. Setting the min value for lift to 1 as anything below that indicates lower chance of it being bought. 

```python
rules=association_rules(frequent_itemsets,metric='lift',min_threshold=1)
rules.sort_values('lift',ascending=False,inplace=True)
rules=rules.reset_index()
rules
```

#### Profusion Removal
from the above table its clear that there are many duplicates so following the below steps to remove the duplicates. 

```python
#Using a custom function which returns a list. 
def to_list(i):
    return(list(i))
#The antecedents and consequents are separately added to a list and then concatenated to form a series.
sortrules=rules.antecedents.apply(to_list) + rules.consequents.apply(to_list)
#The series is then sorted in alphabetical order.
sortrules=sortrules.apply(sorted)
#The series is then saved as a list in rulesets.
rulesets=list(sortrules)
#The eliminate the duplicates the list is converted into a tuple and then it is converted into list.
unique_rulesets=[list(m) for m in set (tuple(i) for i in rulesets)]
#Creating an empty list and then storing the index numbers of the remaining itemsets in to the empty list. 
index_rules=[]
for i in unique_rulesets:
    index_rules.append(rulesets.index(i))
#getting rules without redudancy/profusion.
rules_no_redudancy=rules.iloc[index_rules,:]
#Sorting them on the basis of lift and getting top 10 rules.
rules_no_redudancy.sort_values('lift',ascending=False).head(10)
```
